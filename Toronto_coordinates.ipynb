## LATITUDE AND LONGITUDE OF TORONTO'S POSTAL CODES

#### Loading csv file with Postal codes, Boroughs and Neighborhoods from Toronto


```python
import pandas as pd

df_tor_2 = pd.read_csv("df_tor.csv")

print(df_tor_2.shape)
df_tor_2.head()
```

### Loading coordinates of Toronto's Postal codes from csv file

```python
coord_tor = pd.read_csv("Geospatial_Coordinates.csv")

coord_tor.head()
```

### Renaming Postal Code column

```python
coord_tor.rename({'Postal Code': 'PostalCode'}, axis=1, inplace=True)

coord_tor.head(2)
```

## MERGING THE DATAFRAMES
We use the PostalCode column to merge the two Dataframes

```python
df_tor_f = df_tor_2.merge(coord_tor, on="PostalCode", how="left")
df_tor_f.head()
```
